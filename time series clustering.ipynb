### Impute with sliding window linear interpolation:

```python
# define window size and stride
window_size = 3
stride = 1

# impute missing values with sliding window
for i in range(0, len(df), stride):
    window = df.iloc[i:i+window_size]
    if len(window) == window_size and window.isna().sum().sum() > 0:
        window.interpolate(method='linear',limit=3, inplace=True, limit_direction='both')
```

## K-means clustering

### Find the best distance to use (eval with silhouette score):

```python
from sklearn.metrics import silhouette_score
from tslearn.clustering import TimeSeriesKMeans
import numpy as np
import math

# Transpose the dataframe
df_transposed = df.transpose()

# Convert the transposed dataframe into a numpy array
data = np.array(df_transposed)

# Reshape the numpy array into a 3D array
n_samples, n_features = data.shape
sequence_length = 1
data = data.reshape((n_samples, n_features, sequence_length))

# Perform time series clustering using TimeSeriesKMeans
n_clusters = math.ceil(math.sqrt(len(data))) 

# Calculate silhouette score for "euclidean" distance
km_euclidean = TimeSeriesKMeans(n_clusters=n_clusters, metric="euclidean", verbose=True)
labels_euclidean = km_euclidean.fit_predict(data)
silhouette_euclidean = silhouette_score(data.reshape(n_samples, n_features), labels_euclidean)

# Calculate silhouette score for "dtw" distance
km_dtw = TimeSeriesKMeans(n_clusters=n_clusters, metric="dtw", verbose=True)
labels_dtw = km_dtw.fit_predict(data)
silhouette_dtw = silhouette_score(data.reshape(n_samples, n_features), labels_dtw)

# Calculate silhouette score for "softdtw" distance
km_softdtw = TimeSeriesKMeans(n_clusters=n_clusters, metric="softdtw", verbose=True)
labels_softdtw = km_softdtw.fit_predict(data)
silhouette_softdtw = silhouette_score(data.reshape(n_samples, n_features), labels_softdtw)

print("Silhouette score for 'euclidean':", silhouette_euclidean)
print("Silhouette score for 'dtw':", silhouette_dtw)
print("Silhouette score for 'softdtw':", silhouette_softdtw)
```

```python

km = TimeSeriesKMeans(n_clusters=n_clusters, metric="softdtw", verbose=True, random_state=23122003)
labels = km.fit_predict(data)
```

```python
import math
import matplotlib.pyplot as plt

plot_count = math.ceil(math.sqrt(n_clusters))
som_x = som_y = math.ceil(math.sqrt(math.sqrt(len(data))))

fig, axs = plt.subplots(plot_count,plot_count,figsize=(25,25))
fig.suptitle('Clusters')
row_i=0
column_j=0

colors = plt.cm.jet(np.linspace(0, 1, len(data)))
for label in set(labels):
    cluster = []
    for i in range(len(labels)):
        if(labels[i]==label):
            axs[row_i, column_j].plot(data[i],c=colors[i],alpha=0.4)
            cluster.append(data[i])
            axs[row_i, column_j].set_title("Cluster "+str(row_i*som_y+column_j))
    if len(cluster) > 0:
        axs[row_i, column_j].plot(np.average(np.vstack(cluster),axis=0),c="red")
    column_j+=1
    if column_j%plot_count == 0:
        row_i+=1
        column_j=0
        
plt.show()
```

```python
cluster_c = [len(labels[labels==i]) for i in range(n_clusters)]
cluster_n = ["Cluster "+str(i) for i in range(n_clusters)]
plt.figure(figsize=(15,5))
plt.title("Cluster Distribution for KMeans")
plt.bar(cluster_n,cluster_c)
plt.show()
```
